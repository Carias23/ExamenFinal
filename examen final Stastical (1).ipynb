Identificacion de valores atipicos y almacenando en un dataframe

Método IQR

```python
temp = pd.DataFrame()

for col in data_clean.columns:
    column = data_clean[col]
 
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)

    IQR = Q3 - Q1
    
    umbral_inferior = Q1 - 1.5 * IQR
    umbral_superior = Q3 + 1.5 * IQR

    data_clean[col + '_outliers'] = (column < umbral_inferior) | (column > umbral_superior)

    #temp[col + '_outliers'] = ((column < umbral_inferior) | (column > umbral_superior))

    print(f'{Q1 =} {Q3=} {IQR =} {umbral_inferior = } {umbral_superior =}')
```

Valores transformados mediante Box-Cox

```python
from scipy import stats

lambdas = []
nombre = []
df_new = pd.DataFrame()

for x in data_noout.columns:
    df, lamb = stats.boxcox(data_noout[x])
    lambdas.append(lamb)
    nombre.append(x)
    df_new[x] = df
```

Análisis para determinar el número óptimo de grupos (k) para la agrupación de K-Means utilizando el método del codo

```python
from sklearn.cluster import KMeans
```

```python
inercia = []
ks = []

for k in range(1,60):
    modelo_iterado = KMeans(n_clusters = k, random_state = 5)
    modelo_iterado.fit(data_bc)
    inercia.append(modelo_iterado.inertia_)
    ks.append(k)
```

DataFrame llamado "codo" a partir de dos diccionarios

```python
codo = {
    'inercia': inercia,
    'k': ks
}

codo = pd.DataFrame(codo)
```

Grafico de Dispersion y de Lineas

```python
sns.scatterplot(data=codo,x='k',y='inercia',markers=True)
sns.lineplot(data=codo, x='k', y='inercia',)
```

Modelo de clustering K-Means con 5 clusters y ajustándolo

```python
modelo = KMeans(n_clusters=5,random_state = 10)
modelo.fit(data_bc)
```

Asignando etiquetas de clúster

```python
data_noout['cluster'] = modelo.labels_
data_noout
```
